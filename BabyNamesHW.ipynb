```python
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", '{:,}'.format)
```

```python
import glob

with open("result.txt", "w") as outfile:
    # Loop through each year from 2000 to 2022
    for year in range(2000, 2023):
        # Generate the filename for the current year
        filename = f"yob{year}.txt"
        
        # Check if the file exists before attempting to read it
        if glob.glob(filename):
            with open(filename, "r") as infile:
                for line in infile:
                    # Strip the line to remove any extra whitespace
                    line = line.strip()
                    # Append the year to the line
                    line_with_year = f"{line},{year}"
                    # Write the modified line to the result file
                    outfile.write(line_with_year + "\n")
        else:
            print(f"File {filename} not found.")
```

```python
df = pd.read_csv("result.txt", sep=",", header=None, names=["Name", "Gender", "Count", "Year"])
df.head(10)
```
